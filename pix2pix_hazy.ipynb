# Visualize

```
import numpy as np
import matplotlib.pyplot as plt

w=10
h=10
fig=plt.figure(figsize=(20, 20))
columns = 3
rows = 5
num=2533
p=0
add = ['_real_B.png','_real_A.png','_fake_B.png']
title = ['Real','Hazy','Generated']
for i in range(1, columns*rows +1):
    img = plt.imread('./results/dehazing_pix2pix/test_latest/images/00' + str(num) + add[p])
    a = fig.add_subplot(rows, columns, i)
    a.set_title(title[p])
    a.set_xticklabels([])
    a.set_yticklabels([])
    plt.imshow(img)
    p+=1;
    if(p==3):
      num+=1
      p=0
plt.axis('off')
plt.show()
```
